# Box and violin charts

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## Grouped box plot

```python
# YOUR CODE HERE
df = pd.read_csv("10_OneNumSevCatSubgroupsSevObs.csv")
df['tip_share'] = round(df.tip / df.total_bill * 100, 2)
df_extract = df[['sex', 'day','tip_share']]
```

```python
# YOUR PLOT HERE
sns.set_style('whitegrid')
fig, ax = plt.subplots(figsize=(10,6))
sns.swarmplot(x="sex",y="tip_share",
                data=df_extract,color='lightgray',
                size=5, ax=ax)

sns.boxplot(x='sex', y = 'tip_share', 
                hue = 'day', hue_order=["Thur", "Fri", "Sat", "Sun"], 
                data = df_extract, 
                palette="pastel", ax=ax)

sns.despine(top=True,left=True,right=True)
ax.set(ylim=(0, 40))
ax.set(ylabel="Tip (%)")
```

## Violin plot

```python
# YOUR PLOT HERE
sns.set_style('whitegrid')
colors = ["#FFF19E", ]
g = sns.catplot(x='day',y ='tip_share', 
                hue ='sex',
                order=["Thur", "Fri", "Sat", "Sun"],
                data = df_extract, kind= "violin", 
                palette="pastel", 
                split=True)
g.set_ylabels("Tip (%)")
g.set(ylim=(-10, 100))
```
